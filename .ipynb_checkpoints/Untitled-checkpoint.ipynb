```python
# Positive vs negative word analysis.
```

```python
from collections import Counter
import re

def tokenize(text):
    words = re.findall(r'\b\w+\b', text.lower())
    return words

word_counts = Counter()
for text in d['review']:
    word_counts.update(tokenize(text))

print(word_counts.most_common(20))
```

```python
pos = d[d['sentiment']==1]
neg = d[d['sentiment']==0]
```

```python
pos_counts = Counter()
neg_counts = Counter()

for text in pos['review']:
    pos_counts.update(tokenize(text))

for text in neg['review']:
    neg_counts.update(tokenize(text))
```

```python
print("Positive:", pos_counts.most_common(20))
print("Negative:", neg_counts.most_common(20))
```
